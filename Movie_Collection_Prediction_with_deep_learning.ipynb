Splitting the dataset into test-train data

```python
from sklearn.model_selection import train_test_split
X_train_full, X_test, y_train_full, y_test = train_test_split(data_x, data_y, random_state=42)
X_train, X_valid, y_train, y_valid = train_test_split(X_train_full, y_train_full, random_state=42)
```

Normalisation of the training dataset

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_valid = scaler.transform(X_valid)
X_test = scaler.transform(X_test)
```

```python
X_train
```

```python
np.random.seed(42)
tf.random.set_seed(42)
```

Creating model

```python
model = keras.models.Sequential([
    keras.layers.Dense(30, activation="relu", input_shape=[19]),
    keras.layers.Dense(30, activation="relu"),
    keras.layers.Dense(1)
])
```

```python
model.summary()
```

```python
model.compile(loss="mean_squared_error", ###compiling the model
              optimizer=keras.optimizers.SGD('learning_rate'==1e-2),
              metrics=['mae'])
```

training the model

```python
model_history = model.fit(X_train, y_train, epochs=100, validation_data=(X_valid, y_valid))
```

```python
mae_test = model.evaluate(X_test, y_test)
```
